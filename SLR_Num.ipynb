```python
# Simple Linear Regression Equation from Scratch
# Given data:
# x = 1,2,3,4,5
# y = 1.2,1.8,2.6,3.2,3.8

import numpy as np

# Step 1: Input data
x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([1.2, 1.8, 2.6, 3.2, 3.8], dtype=float)
import numpy as np

# Step 1: Input data
x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([1.2, 1.8, 2.6, 3.2, 3.8], dtype=float)

# Step 2: Mean values
x_mean = np.mean(x)
y_mean = np.mean(y)

# Step 3: Calculate slope (b1)
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)

b1 = numerator / denominator

# Step 4: Calculate intercept (b0)
b0 = y_mean - b1 * x_mean

# Step 5: Print regression equation
print("Slope (b1):", round(b1, 4))
print("Intercept (b0):", round(b0, 4))
print(f"\nEquation: y = {round(b1,4)}x + {round(b0,4)}")

# Step 6: Prediction function (Usage Function)
def predict(x_input):
    """
    Predict y value for a given x using trained linear regression model.
    """
    return b1 * x_input + b0

# Step 7: Predict for a single value
x_test = 6
y_test = predict(x_test)

print(f"\nPrediction:")
print(f"For x = {x_test}, predicted y = {round(y_test, 4)}")

# Step 8: Predict for multiple values
x_new = [6, 7, 8, 9]
y_new = [predict(val) for val in x_new]

print("\nBatch Predictions:")
for i in range(len(x_new)):
    print(f"x = {x_new[i]} → y = {round(y_new[i], 4)}")
# Step 2: Calculate means
x_mean = np.mean(x)
y_mean = np.mean(y)

# Step 3: Calculate slope (b1)
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)

b1 = numerator / denominator

# Step 4: Calculate intercept (b0)
b0 = y_mean - b1 * x_mean

# Step 5: Print regression equation
print("Slope (b1):", round(b1, 4))
print("Intercept (b0):", round(b0, 4))

print(f"\nSimple Linear Regression Equation:")
print(f"y = {round(b1,4)}x + {round(b0,4)}")

# Step 6: Predict values
y_pred = b1 * x + b0

print("\nPredicted values:")
for i in range(len(x)):
    print(f"x = {x[i]}, Predicted y = {round(y_pred[i],4)}")
```
